### Find Examples Used to Check for Bias Based on Size n
This will take the first 10 lower than and first 10 upper than the pre-definied length of text (in this case 50 words long) for both the toxic and non-toxic sample.
Will end up with 40 examples total

```python
# index of the comment_text that will be focused on in this bias analysys
tsld_class0 = []  # all comments below the pre-defined length of 50 words
tsld_class1 = []  # all comments above the pre-defined length of 50 words
ntsld_class0 = []  # all comments below the pre-defined length of 50 words
ntsld_class1 = []  # all comments above the pre-defined length of 50 words

# function to find the length (in words) of some text
def len_str(text):
    l = text.split(' ')
    return len(l)


n = 10

# check for the length less than or equal to 50 in toxic_labaled_data
index = 0
while len(tsld_class0) != n:
    text = toxic_sample_labaled_data.iloc[index][0]
    toxic = toxic_sample_labaled_data.iloc[index][1]
    l = len_str(text)
    if l <= 50:
        tsld_class0.append([text, toxic])
    index += 1

# check for the length greater than 50 in toxic_sample_labaled_data
index = 0
while len(tsld_class1) != n:
    text = toxic_sample_labaled_data.iloc[index][0]
    toxic = toxic_sample_labaled_data.iloc[index][1]
    l = len_str(text)
    if l > 50:
        tsld_class1.append([text, toxic])
    index += 1

    
# check for the length less than or equal to 50 in non_toxic_labaled_data
index = 0
while len(ntsld_class0) != n:
    text = non_toxic_sample_labaled_data.iloc[index][0]
    toxic = non_toxic_sample_labaled_data.iloc[index][1]
    l = len_str(text)
    if l <= 50:
        ntsld_class0.append([text, toxic])
    index += 1

# check for the length greater than 50 in non_toxic_labaled_data
index = 0
while len(ntsld_class1) != n:
    text = non_toxic_sample_labaled_data.iloc[index][0]
    toxic = non_toxic_sample_labaled_data.iloc[index][1]
    l = len_str(text)
    if l > 50:
        ntsld_class1.append([text, toxic])
    index += 1
    
tsld = tsld_class0 + tsld_class1
ntsld = ntsld_class0 + ntsld_class1

for t in tsld:
    print(t)
print()
for t in ntsld:
    print(t)
    
data_list = tsld + ntsld

all_toxic = []
for data in data_list:
    all_toxic.append(data[1])

print(all_toxic)
```

### Perspective API
y-predicted

```python
from sklearn.metrics import accuracy_score

y_predicted = []
threshold = 0.5

for data in data_list:
    text = data[0]
    try:  # try except if "languageNotSupportedByAttributeError" occurs
        value = get_toxicity_score(text)
    except:
        pass
    if value > threshold:
        y_predicted.append(1)
    else:
        y_predicted.append(0)

print(y_predicted)

accuracy = accuracy_score(y_predicted, y_actual)

print(accuracy)  # the accuracy between y_predicted and y_actual for all the data (yes and no for both comment ranges)
```

# Assessing Model Fairness w.r.t toxicity

```python
# extract indices of examples for gender "Male"

# find the thresholds (over or under 50 words) for comments
threshold_column = []
for data in data_list:
    length = len_str(data[0])
    if length > 50:
        threshold_column.append(1)
    else:
        threshold_column.append(0)

over_50_indices = []  # yes indices
under_50_indices = []  # no indices

for i in range(len(threshold_column)):
    if threshold_column[i] == 1:  # if word count is over 50
        over_50_indices.append(i)
    else:
        under_50_indices.append(i)

y_actual_o50 = [y_actual[i] for i in over_50_indices]
y_predicted_o50 = [y_predicted[i] for i in over_50_indices]

y_actual_u50 = [y_actual[i] for i in under_50_indices]
y_predicted_u50 = [y_predicted[i] for i in under_50_indices]
```

### Accuracy

```python
def class_wise_acc(y_actual, y_predicted):
    total_p = 0
    total_n = 0
    TP=0
    TN=0
    for i in range(len(y_predicted)):
        if y_actual[i] == 1:
            total_p += 1
            if y_actual[i]==y_predicted[i]:
                TP += 1
        
        if y_actual[i] == 0:
            total_n += 1
            if y_actual[i]==y_predicted[i]:
                TN += 1
    
    return(TP/total_p, TN/total_n)

class_1_acc_yes, class_0_acc_yes = class_wise_acc(y_actual_o50, y_predicted_o50)
class_1_acc_no, class_0_acc_no = class_wise_acc(y_actual_u50, y_predicted_u50)

print (f"Class 1 (i.e., yes) accuracy for comments over 50 words = {class_1_acc_yes}")
print (f"Class 0 (i.e., yes) accuracy for comments under 50 words = {class_0_acc_yes}")
print (f"Class 1 (i.e., no) accuracy for comments over 50 words = {class_1_acc_no}")
print (f"Class 0 (i.e., no) accuracy for comments under 50 words = {class_0_acc_no}")
```
